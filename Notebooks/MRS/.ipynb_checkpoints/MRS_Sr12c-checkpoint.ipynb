## Notebook for analysis of MIRI-MRS data and spectral extraction

```python
# Standard packages
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from astropy.visualization import simple_norm

import asdf
from scipy import interpolate
from astropy.stats import sigma_clip

# ExoCAT package
from ExoCAT.mrs_tools import load_cube_dir, measure_position_fit, extract_spectrum

system_name = "Sr12c"
```

### 1. Load path and data

```python
path = f'/Users/mmalin/MIRI/MRS/{system_name}/' # general path for the project

# nice colors to plot the bands
colors = {'1A': 'navy','1B': 'royalblue','1C': 'slateblue',
          '2A': 'indigo','2B': 'blueviolet','2C': 'mediumpurple',
          '3A': 'plum','3B': 'violet','3C': 'purple',
          '4A': 'mediumvioletred','4B': 'hotpink','4C': 'lightpink'}
bands = ["1A","1B","1C","2A","2B","2C","3A","3B","3C",'4A','4B','4C']
```

```python
# Load data cubes : 
science = load_cube_dir(path+'/DATA/Obs2/stage3/') # where the data are stored

# Remove Nan and convert into uJy
cubes = {}
for band in bands :
    cubes[band] = np.where(np.isnan(science[band]['cube']), 0, science[band]['cube'])
    cubes[band] *= 1e6 # convert into muJy

# Create directory where to store the results and plots.
output_dir = path+'Results/'
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
```

### 2. Measure the position of the planet

```python
### Measure the position of the planet fitting a 2D gaussian
position_planet = {}
border = 5    # number of pixels to exclude on each side
for band in bands:
    print(band)

    im_median = np.nanmedian(cubes[band], axis=0)
    ny, nx = im_median.shape

    # Create a masked that excludes the outer +/- 5 px (avoid border of the cubes)
    masked_im = np.copy(im_median)
    masked_im[:border, :]      = -np.inf
    masked_im[-border:, :]     = -np.inf
    masked_im[:, :border]      = -np.inf
    masked_im[:, -border:]     = -np.inf

    # Maximum search inside the valid region only
    pos = np.unravel_index(np.nanargmax(masked_im), masked_im.shape)
    pos_max = (pos[1], pos[0])   # convert to (x, y)
    print(pos_max)

    position_planet[band] = measure_position_fit(cubes[band], pos_max, plot=True, window=(5,5))

# manually fix the last band
position_planet['4C'] = position_planet['4A']
```

```python
# PLOT THE CUBES
fig = plt.figure(figsize = (10,10), tight_layout = False)
for ch in range(len(bands)):
    ax = fig.add_subplot(4, 3, 1+ch)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size = "5%", pad = 0.05)
    im_median = np.nanmedian(cubes[bands[ch]], axis=0)
    im = ax.imshow(im_median, cmap="inferno",origin='lower',vmin=0, vmax=1e2)
    cbar = fig.colorbar(im, cax = cax, orientation = "vertical")
    cbar.set_label("Flux [uJy]", fontsize=12, rotation=270, labelpad=15)
    ax.set_title(f'Band {bands[ch]}'+': '+str(np.around(science[bands[ch]]['wav'][0],2))
                 +' - '+str(np.around(science[bands[ch]]['wav'][-1],2))+' $\mu$m')
    #ax.add_artist(plt.Circle(position_planet[bands[ch]], 2, color='lime', fill=False))
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.tight_layout()
plt.savefig(output_dir+'Cubes.pdf')
plt.show()
```

### 3. Extract the spectrum

```python
# reference files to get the aperture correction factors 
path_ref = "/Users/mmalin/crds_cache/references/jwst/miri/jwst_miri_apcorr_0008.asdf" # 4,5,6,7
af = asdf.open(path_ref)
apcorr = af['apcorr_table']['apcorr'] # corresponds to 0.5, 1, 1.5, 2.5, and 3 fwhm
apcorr_fwhm05, apcorr_fwhm1, apcorr_fwhm15,apcorr_fwhm25 = apcorr[0], apcorr[1],apcorr[2], apcorr[3] 
apcorr_radius = af['apcorr_table']['radius']
apcorr_wave = af['apcorr_table']['wavelength'] 
apcorr05, apcorr1, apcorr15,apcorr25, = {},{},{},{}
for band in bands[:-1]:
    apcorr05[band] = interpolate.interp1d(apcorr_wave,apcorr_fwhm05)(science[band]['wav'])
    apcorr1[band] = interpolate.interp1d(apcorr_wave,apcorr_fwhm1)(science[band]['wav'])
    apcorr15[band] = interpolate.interp1d(apcorr_wave,apcorr_fwhm15)(science[band]['wav'])
    apcorr25[band] = interpolate.interp1d(apcorr_wave,apcorr_fwhm25)(science[band]['wav'])
```

```python
# Extract flux : 
flux_fwhm1, flux_fwhm2dot5 =  {}, {}
for band in bands[:-1]:
    flux_ext, psf_size_wave = extract_spectrum(cubes[band], science[band]['wav'], position_planet[band], science[band]['pxsc'], fwhm_size=1)
    flux_fwhm1[band] = sigma_clip(flux_ext, sigma=10).data * apcorr1[band] # clean for outliers 

    flux_ext, psf_size_wave = extract_spectrum(cubes[band], science[band]['wav'], position_planet[band], science[band]['pxsc'], fwhm_size=2.5)
    flux_fwhm2dot5[band] = sigma_clip(flux_ext, sigma=10).data * apcorr25[band]
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,5))
for band in bands[:-1]:
    ax.plot(science[band]['wav'], flux_fwhm1[band], color=colors[band],label=f' {band}',alpha=0.8)
ax.legend(fontsize=11,ncol=4)
ax.grid(alpha=0.3)
ax.set_xlabel('Wavelength ($\mu$m)', size=15)
ax.set_ylabel('Flux ($\mu$Jy)', size=15)
ax.set_xlim(4.8,25)
#ax.set_yscale('log')
ax.set_ylim(0,2.5e3)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
fig.savefig(output_dir+f'/Spectrum_{system_name}.pdf',bbox_inches='tight', pad_inches=0.1)
plt.show()
```

### 4. Uncertainties

```python
sectors = [f"{i}{l}" for i in range(1, 5) for l in "ABC"]
sectors.remove("4C")  # since 4C is not present
angles_rad = np.radians(np.arange(0, 361, 30))  # apertures angle ; e.g. every 30deg
angles = {s: angles_rad for s in sectors}
```

```python
separation, coord_err, flux_err,flux_apertures,flux_err_tot,err_pipeline = {},{}, {},{},{},{}
for band in bands[:-1]:
    sep_pix =  8
    separation[band] = np.sqrt(sep_pix**2+sep_pix**2)
    
    # Calcul des coordonnes cartsiennes autour du point fixe
    x = position_planet[band][0] + separation[band] * np.cos(angles[band])
    y = position_planet[band][1] + separation[band] * np.sin(angles[band])
    coord_err[band] = np.array([x,y])

    # for each position 
    flux_apertures[band] = np.ndarray((len(angles[band]),len(science[band]['wav'])))
    for i in range(len(angles[band])):
        flux_apertures[band][i,:], psf_size_wave = extract_spectrum(cubes[band], science[band]['wav'],
                                                   (coord_err[band][0][i],coord_err[band][1][i]),
                                                   science[band]['pxsc'], fwhm_size=1)

    # standard dev at each wavelengths :
    flux_err[band] = np.nanstd(flux_apertures[band], axis=0)
   
    # Error from the pipeline :
    err_pipeline[band], psf_size_wave = extract_spectrum(science[band]['err']*1e6, science[band]['wav'],
                                                         position_planet[band],science[band]['pxsc'], fwhm_size=1)

    flux_err_tot[band] = np.sqrt(err_pipeline[band]**2+flux_err[band]**2)
```

```python
# Plot the cube
fig = plt.figure(figsize = (10,10), tight_layout = False)
for ch in range(len(bands)-1):
    ax = fig.add_subplot(4, 3, 1+ch)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size = "5%", pad = 0.05)
    im_median = np.nanmedian(cubes[bands[ch]], axis=0)
    norm = simple_norm(im_median, stretch='asinh', vmin=1, vmax=1e2)
    im = ax.imshow(im_median, cmap="inferno",origin='lower',norm=norm)
    cbar = fig.colorbar(im, cax = cax, orientation = "vertical")
    ax.set_title('Band '+bands[ch])
    for i in range(len(angles[bands[ch]])):
        ax.scatter(coord_err[bands[ch]][0][i],coord_err[bands[ch]][1][i], color='white', marker='P')
    ax.scatter(position_planet[bands[ch]][0],position_planet[bands[ch]][1], color='dodgerblue', marker='P')
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.tight_layout()
plt.show()
```

```python
flux_fwhm1, flux_fwhm2dot5 =  {}, {}
for band in bands[:-1]:
    flux_ext, psf_size_wave = extract_spectrum(cubes[band], science[band]['wav'], position_planet[band], 
                                               science[band]['pxsc'], fwhm_size=1)
    flux_fwhm1[band] = sigma_clip(flux_ext, sigma=10).data * apcorr1[band]

    flux_ext, psf_size_wave = extract_spectrum(cubes[band], science[band]['wav'], position_planet[band],
                                               science[band]['pxsc'], fwhm_size=2.5)
    flux_fwhm2dot5[band] = sigma_clip(flux_ext, sigma=10).data * apcorr25[band]
    
    spec_to_export = pd.DataFrame(np.vstack((science[band]['wav'], flux_fwhm1[band], flux_err_tot[band])).T)
    spec_to_export.to_csv(output_dir+f'/Spectrum_{system_name}_{band}_1fwhm.txt', header=['Wavelength[um]', 'Flux [uJy]', 'Err [uJy]'], 
                          sep = '\t', index=False)

    spec_to_export = pd.DataFrame(np.vstack((science[band]['wav'], flux_fwhm2dot5[band], flux_err_tot[band])).T)
    spec_to_export.to_csv(output_dir+f'/Spectrum_{system_name}_{band}_2dot5fwhm.txt', header=['Wavelength[um]', 'Flux [uJy]',  'Err [uJy]'], 
                          sep = '\t', index=False)
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,5))
for band in bands[:-1]:
    
    ax.errorbar(science[band]['wav'], flux_fwhm1[band], yerr=flux_err_tot[band], color=colors[band],alpha=0.05)
    ax.plot(science[band]['wav'], flux_fwhm1[band], color=colors[band],label=f' {band}',alpha=1)

ax.legend(fontsize=11,ncol=4)
ax.grid(alpha=0.3)
ax.set_xlabel('Wavelength ($\mu$m)', size=15)
ax.set_ylabel('Flux ($\mu$Jy)', size=15)
ax.set_xlim(4.8,25)
ax.set_ylim(1,2.5e3)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
fig.savefig(output_dir+f'/Spectrum_{system_name}.pdf',bbox_inches='tight', pad_inches=0.1)
plt.show()
```
